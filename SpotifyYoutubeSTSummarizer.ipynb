### Necessary imports 🔧




```python
import torch
from transformers import AutoModelForSpeechSeq2Seq, AutoProcessor, pipeline
```

### Define Model checkpoint, device and datatype 🔉

```python
device = "cuda:0" if torch.cuda.is_available() else "cpu"
torch_dtype = torch.float16 if torch.cuda.is_available() else torch.float32

model_id = "distil-whisper/distil-large-v3"

model = AutoModelForSpeechSeq2Seq.from_pretrained(
    model_id, torch_dtype=torch_dtype, low_cpu_mem_usage=True, use_safetensors=True, attn_implementation="sdpa"
)
model.to(device)
```

### Load the model and initialise the speech recognition pipeline ⚡

```python
processor = AutoProcessor.from_pretrained(model_id)

pipe = pipeline(
    "automatic-speech-recognition",
    model=model,
    tokenizer=processor.tokenizer,
    feature_extractor=processor.feature_extractor,
    max_new_tokens=128,
    torch_dtype=torch_dtype,
    device=device,
)
```

```python
from fastapi import FastAPI, Form
from fastapi.responses import HTMLResponse
import uvicorn
from pyngrok import ngrok
import yt_dlp
import whisper
from transformers import pipeline
import torch
import re
import nest_asyncio
import os
import glob
import subprocess

app = FastAPI()

# Load Whisper model
whisper_model = whisper.load_model("base")

# Set up the summarization pipeline
device = "cuda" if torch.cuda.is_available() else "cpu"
summarizer = pipeline("summarization", model="facebook/bart-large-cnn", device=device)

def is_youtube_link(url):
    youtube_regex = (
        r'(https?://)?(www\.)?'
        '(youtube|youtu|youtube-nocookie)\.(com|be)/'
        '(watch\?v=|embed/|v/|.+\?v=)?([^&=%\?]{11})')
    return re.match(youtube_regex, url) is not None

def is_spotify_link(url):
    spotify_regex = r'https?://open\.spotify\.com/track/[a-zA-Z0-9]+'
    return re.match(spotify_regex, url) is not None

def download_youtube_audio(url):
    ydl_opts = {
        'format': 'bestaudio/best',
        'postprocessors': [{
            'key': 'FFmpegExtractAudio',
            'preferredcodec': 'mp3',
            'preferredquality': '192',
        }],
        'outtmpl': 'audio.%(ext)s',
    }
    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        ydl.download([url])

def download_spotify_audio(url):
    # Delete any existing MP3 files in the current directory
    existing_files = glob.glob('*.mp3')
    for file in existing_files:
        os.remove(file)
    print('Deleted existing MP3 files in current directory\n')

    # Download the song using spotdl
    print('Downloading song from Spotify\n')
    subprocess.run(['spotdl', url, '--audio', 'youtube-music', '--format', 'mp3', '--bitrate', '320k', '--output', '.'])
    print('\nDownloaded song is located in current directory')

    # Rename the downloaded file to audio.mp3
    downloaded_files = glob.glob('*.mp3')
    if downloaded_files:
        os.rename(downloaded_files[0], 'audio.mp3')

def transcribe_audio():
    result = whisper_model.transcribe("audio.mp3")
    return result["text"]

def summarize_text(text):
    summary = summarizer(text, max_length=150, min_length=50, do_sample=False)[0]['summary_text']
    return summary

@app.get("/", response_class=HTMLResponse)
async def read_root():
    return """
    <html>
        <body>
            <h1>Enter a YouTube or Spotify link</h1>
            <form action="/process" method="post">
                <input type="text" name="url" />
                <input type="submit" />
            </form>
        </body>
    </html>
    """

@app.post("/process", response_class=HTMLResponse)
async def process_link(url: str = Form(...)):
    try:
        if is_youtube_link(url):
            link_type = "YouTube"
            download_youtube_audio(url)
        elif is_spotify_link(url):
            link_type = "Spotify"
            download_spotify_audio(url)
        else:
            return "Invalid link. Please provide a valid YouTube or Spotify link."

        transcription = transcribe_audio()
        summary = summarize_text(transcription)

        return f"""
        <html>
            <body>
                <h1>Results</h1>
                <h2>Link Type: {link_type}</h2>
                <h3>Transcription:</h3>
                <p>{transcription}</p>
                <h3>Summary:</h3>
                <p>{summary}</p>
            </body>
        </html>
        """
    except Exception as e:
        return f"""
        <html>
            <body>
                <h1>Error</h1>
                <p>An error occurred while processing your request: {str(e)}</p>
            </body>
        </html>
        """

# Set up ngrok
ngrok_tunnel = ngrok.connect(8000)
print('Public URL:', ngrok_tunnel.public_url)

# Run the FastAPI app
nest_asyncio.apply()

# Now you can run uvicorn without encountering the RuntimeError
uvicorn.run(app, port=8000)
```
